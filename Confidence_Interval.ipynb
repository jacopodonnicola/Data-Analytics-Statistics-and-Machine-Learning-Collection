```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Mean and standard deviation of the normal distribution
mean = 12
std_dev = 2.7688746209726918

# Generate data points for the x-axis
x = np.linspace(mean - 4*std_dev, mean + 4*std_dev, 100)

# Calculate the probability density function (PDF) for the normal distribution
pdf = norm.pdf(x, mean, std_dev)

# Plot the normal distribution
plt.plot(x, pdf, color='blue')
plt.xlabel('')
plt.ylabel('Probability Density')
plt.title('Normal Distribution')

# Confidence interval values
lower_bound = 9.784438872179079
upper_bound = 14.215561127820921

# Plot vertical lines for the confidence interval
plt.axvline(x=lower_bound, color='red', linestyle='--', label='Lower Bound (95%)')
plt.axvline(x=upper_bound, color='red', linestyle='--', label='Upper Bound (95%)')

# Plot a symbol for the mean
plt.axvline(x=mean, color='black', linestyle='-', label='Mean = 12')

plt.legend()
plt.grid(True)
plt.show()
```
