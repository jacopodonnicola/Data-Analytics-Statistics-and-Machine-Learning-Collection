## Investigation on SQL

The dataset contains some null values and empty fields. During this initial stage of exploration, they will be excluded from analysis as necessary, directly within the query script

#### How many riders, drivers and taxi are in the dataset?

```python
%%sql 
SELECT 
COUNT(DISTINCT user_id) AS Tot_riders,
COUNT(DISTINCT driver_id) AS Tot_drivers,
COUNT(DISTINCT taxi_id) AS Tot_taxi
FROM taxi;
```

#### What is the ratio between drivers and riders ?

```python
%%sql 

WITH t1 AS (
SELECT 
COUNT(DISTINCT user_id) AS Tot_riders,
COUNT(DISTINCT driver_id) AS Tot_drivers
FROM taxi
)
SELECT
ROUND(CAST(Tot_drivers AS FLOAT) / Tot_riders,2) AS Ratio_Between_Drivers_and_Riders

FROM t1 
```

#### Which are the different vehicle types, how much journeys and percentage correspond to each one ?

```python
%%sql 

SELECT
DISTINCT icon AS Service_type, 
COUNT(icon) AS Num_of_journeys, 
ROUND((COUNT(icon) * 100.0 / (SELECT COUNT(*) FROM taxi)), 2) AS Percentage
FROM taxi
GROUP BY 1
```

#### Which are the different request types and how much journeys and percentage correspond to each one?

```python
%%sql 

SELECT
DISTINCT start_type AS Request_type, count(icon) AS Num_of_journeys,
ROUND((COUNT(icon) * 100.0 / (SELECT COUNT(*) FROM taxi)), 2) AS Percentage
FROM taxi
GROUP BY 1
```

#### How the requests end and how much journeys and percentage correspond to each one?

```python
%%sql 

SELECT
DISTINCT end_state AS Request_outcome, 
COUNT(icon) AS Num_of_journeys,
ROUND((COUNT(icon) * 100.0 / (SELECT COUNT(*) FROM taxi)), 2) AS Percentage
FROM taxi
WHERE Request_outcome != 'None'
GROUP BY 1
```

    Conclusions: 17417 requests, corresponding to 75% of the overall, were successfully concluded. 

### ON RIDERS

#### How many journeys has the best 10 riders done, what is the total distance traveled, the duration of the travel, and how much they has been earning?

```python
%%sql 

SELECT user_id AS Rider, 
COUNT(journey_id) AS Num_of_journeys,
SUM(distance) AS Tot_distance,
SUM(duration) AS Tot_duration,
SUM(price) AS Tot_price
FROM taxi 
GROUP BY 1 
ORDER BY 2 DESC
LIMIT 10
```

#### Which 10 passengers used the service the longest?

```python
%%sql 

With Duration AS(
SELECT
user_id,
ABS((strftime('%s', end_at) - strftime('%s', start_at)) / 60) AS Duration_in_minutes
FROM taxi
)

SELECT 
user_id AS Rider,
SUM(Duration_in_minutes) AS Tot_use_in_minutes
FROM Duration
GROUP BY 1
ORDER BY 2 DESC
LIMIT 10;
```

#### How many minutes in average a rider wait for a driver?

```python
%%sql

SELECT
CAST(AVG(Waiting_in_minutes)AS INTEGER) AS Average_time_in_minutes_waited_for_a_driver
FROM (SELECT 
      ABS((strftime('%s', arrived_at) - strftime('%s', start_at)) / 60) AS Waiting_in_minutes 
      FROM taxi) Waiting_rides
```

    Conclusions: The Average time waited for a driver is 10 minutes. We also discovered earlier that for every 10 riders in attendance, on average one driver is available and this affects waiting time. Is this a desired result or could we consider increasing taxi coverage? 

#### How many riders there are in each ranking band?

```python
%%sql 

WITH Scoring AS (
SELECT 
user_id AS Rider,
ROUND(AVG(rider_score),2) AS Average_Score
FROM taxi 
WHERE rider_score IS NOT null
GROUP BY 1
)
SELECT
COUNT(CASE WHEN Average_Score BETWEEN 4 AND 5 THEN 1 END) AS Score_4_5,
COUNT(CASE WHEN Average_Score BETWEEN 3 AND 4 THEN 1 END) AS Score_3_4,
COUNT(CASE WHEN Average_Score BETWEEN 2 AND 3 THEN 1 END) AS Score_2_3,
COUNT(CASE WHEN Average_Score BETWEEN 1 AND 2 THEN 1 END) AS Score_1_2,
COUNT(CASE WHEN Average_Score BETWEEN 0 AND 1 THEN 1 END) AS Score_0_1
FROM Scoring
```

    Concusions: We can see that the majority of users are rated positively on average. However, there are a few users who have a very low rating. We could check out the possible reasons for this.

### ON DRIVERS

#### Which are the 15 drivers who have made the most drives?

```python
%%sql

SELECT driver_id AS Driver, 
COUNT(journey_id) AS Num_of_journeys, 
SUM(distance) AS Tot_distance,
SUM(duration) AS Tot_duration,
SUM(cost) AS Tot_cost
FROM taxi 
WHERE driver_id != 'None' 
GROUP BY 1 
ORDER BY 2 DESC
LIMIT 15
```

#### Which are in average the 5 fastest drivers to fullfill a request with success from a rider?

```python
%%sql 

SELECT
driver_id AS Driver,
CAST(AVG(ABS((strftime('%s', arrived_at) - strftime('%s', start_at))))AS INTEGER) AS Avg_waited_time_in_seconds
FROM taxi 
WHERE end_state = 'drop off' AND start_type = 'asap'
GROUP BY 1
ORDER BY 2
LIMIT 5;
```

#### Which are in average the 5 slowest drivers to fullfill a request with success from a rider?

```python
%%sql

SELECT
driver_id AS Driver,
CAST(AVG(ABS((strftime('%s', arrived_at) - strftime('%s', start_at))/ 60))AS INTEGER) AS Avg_waited_time_in_minutes
FROM taxi 
WHERE end_state = 'drop off' AND start_type = 'asap'
GROUP BY 1
ORDER BY 2 DESC
LIMIT 5;
```

#### What is the ratio of rides that were not successfully completed?

```python
%%sql 

SELECT
ROUND(CAST(COUNT(CASE WHEN end_state IN 
('driver cancel', 'failure', 'no show', 'not found', 'rider cancel') THEN 1 END) AS FLOAT)/
COUNT(end_state)*100,2) AS Percentage_uncomplete_requests 
FROM taxi
```

#### Which drivers cancel most ride requests?

```python
%%sql

WITH t1 AS (
SELECT
driver_id AS Driver,
COUNT(journey_id) as Tot_rides_cancelled_by_the_driver,
ROUND((COUNT(journey_id) * 100.0 / (SELECT COUNT(journey_id) FROM taxi WHERE end_state = 'driver cancel')),2) AS Percent_of_total
FROM taxi
WHERE end_state = 'driver cancel'
GROUP BY 1
ORDER BY Tot_rides_cancelled_by_the_driver DESC
) 
SELECT *,
ROUND(SUM(Percent_of_total) OVER (ORDER BY Tot_rides_cancelled_by_the_driver DESC),2) AS Progressive_sum_of_percentage,
RANK() OVER (ORDER BY Tot_rides_cancelled_by_the_driver DESC) AS Rank
FROM t1
LIMIT 15
```

    Conclusions: While it's evident that 8 drivers alone are responsible for 46% of cancelled driver requests, it's worth noting that these drivers also account for a significant portion of the total rides completed. To provide a more relative perspective, we can calculate a ratio that considers the proportion of cancelled requests relative to the total number of rides undertaken by each driver.

```python
%%sql 

SELECT
driver_id AS Driver,
ROUND(CAST(COUNT(CASE WHEN end_state = 'driver cancel' THEN 1 END) AS FLOAT)/COUNT(journey_id),2) AS Driver_ride_cancelation_ratio
FROM taxi
WHERE driver_id is not null
GROUP BY 1
ORDER BY Driver_ride_cancelation_ratio DESC
LIMIT 10
```

    Conclusions: Now that we know which drivers most frequently suppress requests, we can deepen the analysis by researching how other variables are related to this result.

```python
%%sql 

SELECT
driver_id AS Driver,
journey_id AS Ride,
icon AS Service_Type,
start_type AS Request_Type,
end_state AS Request_outcome
FROM taxi 
WHERE driver_id = 'ccfcfc81abe6485b505ba742749c8d2b' or driver_id = '8a122c8845dc5eebe7d584eb30428057'
```

    Conclusions: We have identified that the two drivers handled group and executive requests, respectively. Notably, all requests deleted by these drivers were 'asap' requests. This indicates a potential opportunity to optimize coverage strategies to better accommodate 'asap' requests.

#### Which drivers have the best ranking in terms of driver score?

```python
%%sql

WITH Scoring AS (
SELECT
driver_id AS Driver,
ROUND(AVG(driver_score),2) AS Average_Score
FROM taxi
WHERE driver_score is not null
GROUP BY 1 
),
Ranking AS (
SELECT
Driver,
Average_Score, 
DENSE_RANK() OVER (ORDER BY Average_Score DESC) AS Ranking
FROM Scoring
)
SELECT 
Driver,
Average_Score
FROM Ranking
WHERE Ranking = 1
```

    Conclusions: These 13 drivers over 168 have the best ranking in terms of customer satisfaction.

#### Which drivers have the average ranking in terms of driver score between 1 and 2?

```python
%%sql

WITH Scoring AS (
SELECT
driver_id AS Driver,
ROUND(AVG(driver_score),2) AS Average_Score
FROM taxi
WHERE driver_score is not null
GROUP BY 1 
)
SELECT
Driver,
Average_Score
FROM Scoring
WHERE Average_Score >= 1 and Average_Score <= 2
ORDER BY Average_Score DESC
```

    Conclusions: Theare are 6 drivers with a customer satisfaction score between 1 and 2

### ON THE SERVICE

#### How do the different devices perform in relation to the number of requests satisfied?

```python
%%sql 

SELECT
source AS Device,
COUNT(CASE WHEN end_state = 'failure' THEN 1 END) AS Num_of_failed_requests, 
COUNT(end_state) AS Tot_request,
ROUND(CAST(COUNT(CASE WHEN end_state = 'failure' THEN 1 END) AS FLOAT) / COUNT(end_state) * 100, 3) AS Percentage_failed_requests
FROM taxi
WHERE source IS NOT NULL
GROUP BY 1
ORDER BY 4 DESC;
```

    Conclusions: It seems that ride requests from Android devices have the highest failure rate. Could it be due to some technical problem with the application on Android?

#### Is the type of request (asap, reserved, delayed) correlated with the number of failures in running requests?

```python
%%sql 

SELECT 
source AS Device,
start_type AS Request_type,
COUNT(start_type) AS Num_of_requests,
ROUND((COUNT(start_type) * 100.0 / (SELECT COUNT(*) FROM taxi WHERE source = source)), 2) AS Percentage_device_for_request_type
FROM taxi t
WHERE Device is not null
GROUP BY 1,2
```

    Conclusions:  We can state with sufficient certainty that the type of request is not the reason for the number of failed unit requests. Indeed, the iPad device, for exaple, has a proportionally equal number of asap and reserved unit requests, as with the web, but a much higher percentage of failed requests than the web.

### Duration Distribution 

To remove outliers or noise from the data caused by exceptional circumstances, we can focus on the central portion of the data distribution. By considering data ranging from the 2nd to the 98th percentile, we exclude both the smallest values, representing less than 2% of the data, and the largest values, representing the top 2%. This approach helps us to analyze the majority of the data while mitigating the influence of extreme values.

```python
duration_ride = (abs(cleaned_df['end_at']- cleaned_df['arrived_at']).dt.total_seconds()/60).tolist()

p2 = np.percentile(duration_ride, 2)
p98 = np.percentile(duration_ride, 98)

duration_ride_less_noise = [x for x in duration_ride if p2 <= x <= p98]
```

```python
import matplotlib.pyplot as plt

plt.hist(duration_ride_less_noise, bins='auto', color='white', edgecolor='black') 
plt.xlabel('Duration Ride')
plt.ylabel('Number of completed rides')
plt.title('Distribution of Duration Rides')
plt.show()
```

```python
mean = np.mean(duration_ride_less_noise)
mean
```

It is now investigated whether there is a relationship between the total distance travelled by a taxi (executive type) and the revenue generated. This is a query preview, limited to 10 lines, intended to offer an overview of the dataframe used for subsequent analysis. Dimensional correction factors have been applied to ensure the reliability of metrics related to price and distance fields, thus providing a more accurate reflection of real-world values.

```python
%%sql 

WITH t1 AS ( 
SELECT
taxi_id,
ROUND(SUM(price)/800,0) AS Tot_earning,
ROUND(SUM(distance)/10000, 0) AS Tot_distance_traveled
FROM taxi
WHERE end_state = 'drop off' AND icon = 'executive'
GROUP BY 1
)
SELECT
* 
FROM t1
WHERE Tot_earning != 0 and Tot_distance_traveled != 0
LIMIT 10
```

#### The query extracts the total earnings and total distance traveled by each taxi for completed trips (executive type).

```python
conn = sqlite3.connect('Taxi_2010.db')

sql_query =  """

WITH t1 AS ( 
SELECT
taxi_id,
ROUND(SUM(price)/800,0) AS Tot_earning,
ROUND(SUM(distance)/10000, 0) AS Tot_distance_traveled
FROM taxi
WHERE end_state = 'drop off' AND icon = 'executive'
GROUP BY 1
)
SELECT
* 
FROM t1
WHERE Tot_earning != 0 and Tot_distance_traveled != 0

"""

df_query = pd.read_sql_query(sql_query,conn)
```

#### To filter out improbable or incorrect data, extreme values are excluded.

```python
p5 = df_query['Tot_distance_traveled'].quantile(0.05)
p95 = df_query['Tot_distance_traveled'].quantile(0.95)
```

```python
df_query_less_noise = df_query[(df_query['Tot_distance_traveled'] >= p5) & (df_query['Tot_distance_traveled'] <= p95)]
```

```python
plt.scatter(df_query_less_noise['Tot_distance_traveled'] ,df_query_less_noise['Tot_earning'], color='red', label='Data Points')
plt.xlabel('Tot_distance_traveled [Km]')
plt.ylabel('Tot_earning [PEN]') 
plt.grid(True)
```

    Conclusions: The relationship between total revenue and distance travelled by each taxi for trips completed using the executive taxi is strongly linear. This result supports what was observed in tableau, where price was observed as a function of distance travelled on a ride.

### Linear Regression Algorithm

    By identifying a linear correlation between the distance traveled and the earnings, we can leverage machine learning techniques to create a predictive model using the available data. In this case, we can use a simple linear regression algorithm, which aims to model the relationship between the input variable (distance traveled) and the output variable (earnings) by fitting a linear equation to the observed data points. Once the model is trained on the data, it can be used to predict earnings for new instances based on the distance traveled, allowing us to make informed decisions and gain insights into the revenue generation process.

```python
from sklearn.linear_model import LinearRegression
```

```python
X = df_query_less_noise[['Tot_distance_traveled']]  
y = df_query_less_noise['Tot_earning']
```

```python
model = LinearRegression()
```

```python
model.fit(X.values, y)
```

```python
distance_500Km = 500
```

```python
predicted_earnings = model.predict([[distance_500Km]])
predicted_earnings_value = predicted_earnings[0]
predicted_earnings_value
```

```python
plt.scatter(df_query_less_noise['Tot_distance_traveled'] ,df_query_less_noise['Tot_earning'], color='red', label='Data Points')
plt.xlabel('Tot_distance_traveled [Km]')
plt.ylabel('Tot_earnings [PEN]') 
plt.grid(True)

regression_line = np.linspace(min(df_query_less_noise['Tot_distance_traveled']), max(df_query_less_noise['Tot_distance_traveled']), 100).reshape(-1, 1)
predicted_earnings = model.predict(regression_line)
plt.plot(regression_line, predicted_earnings, color='blue', label='Regression Line')

plt.legend()
plt.show()
```

```python
plt.scatter(df_query_less_noise['Tot_distance_traveled'] ,df_query_less_noise['Tot_earning'], color='red', label='Data Points')
plt.scatter(distance_500Km, predicted_earnings_value, color='green', label='Predicted Earnings for 500Km Covered')
plt.xlabel('Tot_distance_traveled [Km]')
plt.ylabel('Tot_earnings [PEN]') 
plt.grid(True)

plt.legend()
plt.show()
```

    Conclusions: We can predict that when a taxi travels 500 km, it will be earning approximately 2564 PEN
