dependent variable = amount spent for week on food

indipendent variable = cost-saving program

H0: the program did not change the cost of food 

HA: the program reduced the cost of food

Ho: mu_program >= mu

HA: mu_program < mu

```python
mu = 151
```

```python
std_dev = 50
```

```python
n = 25
```

```python
std_error = std_dev / math.sqrt(25)
std_error
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import t

# Define the parameters
mean = 151
std_error = 10
df = 24  # Degrees of freedom

# Generate data points
x = np.linspace(mean - 4*std_dev, mean + 4*std_dev, 100)
y = t.pdf(x, df, loc=mean, scale=std_dev)

# Add vertical lines for lower and upper limits
lower_limit = mean - std_dev
upper_limit = mean + std_dev
plt.axvline(lower_limit, color='red', linestyle='--', label='Lower Limit')
plt.axvline(upper_limit, color='red', linestyle='--', label='Upper Limit')

# Plot the t-distribution
plt.plot(x, y, color='blue', label='t-distribution')
plt.xlabel('Value')
plt.ylabel('Probability Density')
plt.title('t-Distribution with Mean = {} and Standard Deviation = {}'.format(mean, std_dev))
plt.legend()
plt.grid(True)
plt.show()
```

### t_Test

```python
mean_sample = 126
```

```python
mean_difference = mean_sample - mu
mean_difference
```

```python
t_Test = mean_difference / std_error
t_Test
```

```python
alpha = 0.05
```

```python
t_critical = -1.711
```

t-Test value is lower than t-crital so the H0 is refused

The p-value is less than 0.05 

```python
D = mean_difference / std_dev
D
```

```python
r_square = t_Test**2 / (t_Test**2 + (n-1))
r_square
```

For the sample of 25 units, 20% of the difference is due to the cost saving program

### Margin of error 

M = t_critical x standard error

```python
M = 2.064 * std_error
M
```

### CI (for the mean)

CI = (mean_sample - M , mean_sample + M)

```python
lower_limit = mean_sample - M 
upper_limit = mean_sample + M
```

```python
CI = (lower_limit, upper_limit)
CI
```
